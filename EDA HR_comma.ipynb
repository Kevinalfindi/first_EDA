```python
#mencari outlier average montly hours

# cari nilai minimum 
p0 = df2.average_montly_hours.min()

# cari nilai maksimum
p100 = df2.average_montly_hours.max()

# cari quartile 1
q1 = df2.average_montly_hours.quantile(0.25)

# cari quartile 2
q2 = df2.average_montly_hours.quantile(0.50)

# cari quartile 3
q3 = df2.average_montly_hours.quantile(0.75)

# cari IQR
iqr= q3-q1

# cari batas bawah
lower_limit= q1-1.5*iqr

# cari batas atas
upper_limit= q3+1.5*iqr

print("batas bawah 'average_montly_hours':", lower_limit)
print("nilai minimum", p0)

if lower_limit <p0:
    print("tidak ada outlier dari sisi bawah")
else: 
    print("ada outlier dari sisi bawah")
print()
print("batas atas 'average_montly_hours':", upper_limit)
print("nilai maksimum", p100)

if upper_limit > p100:
    print("tidak ada outlier dari sisi atas")
else:
    print("ada outlier dari sisi atas")
print()


```

```python
#mencari outlier average montly hours

# cari nilai minimum 
p0 = df2.time_spend_company.min()

# cari nilai maksimum
p100 = df2.time_spend_company.max()

# cari quartile 1
q1 = df2.time_spend_company.quantile(0.25)

# cari quartile 2
q2 = df2.time_spend_company.quantile(0.50)

# cari quartile 3
q3 = df2.time_spend_company.quantile(0.75)

# cari IQR
iqr= q3-q1

# cari batas bawah
lower_limit= q1-1.5*iqr

# cari batas atas
upper_limit= q3+1.5*iqr

print("batas bawah 'time_spend_company':", lower_limit)
print("nilai minimum", p0)

if lower_limit <p0:
    print("tidak ada outlier dari sisi bawah")
else: 
    print("ada outlier dari sisi bawah")
print()
print("batas atas 'time_spend_company':", upper_limit)
print("nilai maksimum", p100)

if upper_limit > p100:
    print("tidak ada outlier dari sisi atas")
else:
    print("ada outlier dari sisi atas")
print()


```
